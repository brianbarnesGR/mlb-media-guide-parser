```python
import pandas as pd
from bs4 import BeautifulSoup
from nameparser import HumanName
import requests
```

# 2018 - Present

```python
team = 'KCR'

_source = {
    2023: 'https://web.archive.org/web/20230314222348/https://www.mlb.com/royals/team/front-office',
    2022: 'https://web.archive.org/web/20220322190146/https://www.mlb.com/royals/team/front-office',
    2021: 'https://web.archive.org/web/20210331133741/https://www.mlb.com/royals/team/front-office'
}
```

```python
final_df = pd.DataFrame()
front_office_directory = dict()

for year in _source.keys():
    front_office_directory[year] = pd.read_html(_source[year])
```

```python
employees = {
    2023: {
        'Senior Administration': front_office_directory[2023][0],
        'Officers': front_office_directory[2023][1],
        'Baseball Operations': front_office_directory[2023][2],
        'Major League Medical Staff': front_office_directory[2023][3],
        'Major League Support Staff': front_office_directory[2023][4],
        'Business Operations': front_office_directory[2023][5],
        'Technology': front_office_directory[2023][6],
        'Ticket Operations': front_office_directory[2023][7],
        'Communications & Broadcasting': front_office_directory[2023][8],
        'Pine Tar Collective': front_office_directory[2023][9],
        'Corporate Partnerships & Client Services': front_office_directory[2023][10],
        'Ticket Sales & Services': front_office_directory[2023][11],
        'Ballpark Operations & Development': front_office_directory[2023][12],
        'Event Operations & Guest Experience': front_office_directory[2023][13],
        'Groundskeeping & Landscaping': front_office_directory[2023][14],
        'Community Impact': front_office_directory[2023][15],
        'Major League Baseball': front_office_directory[2023][16]
    },
    2022: {
        'Senior Administration': front_office_directory[2022][0],
        'Officers': front_office_directory[2022][1],
        'Baseball Operations': front_office_directory[2022][2],
        'Major League Medical Staff': front_office_directory[2022][3],
        'Major League Support Staff': front_office_directory[2022][4],
        'Business Operations': front_office_directory[2022][5],
        'Technology': front_office_directory[2022][6],
        'Ticket Operations': front_office_directory[2022][7],
        'Communications & Broadcasting': front_office_directory[2022][8],
        'Pine Tar Collective': front_office_directory[2022][9],
        'Corporate Partnerships & Client Services': front_office_directory[2022][10],
        'Ticket Sales & Services': front_office_directory[2022][11],
        'Ballpark Operations & Development': front_office_directory[2022][12],
        'Event Operations & Guest Experience': front_office_directory[2022][13],
        'Groundskeeping & Landscaping': front_office_directory[2022][14],
        'Community Impact': front_office_directory[2022][15],
        'Major League Baseball': front_office_directory[2022][16]
    },
    2021: {
        'Ownership': front_office_directory[2022][0],
        'Officers': front_office_directory[2022][1],
        'Baseball Operations': front_office_directory[2022][2],
        'Major League Support Staff': front_office_directory[2022][3],
        'Major League Medical Staff': front_office_directory[2022][4],
        'Business Operations': front_office_directory[2022][5],
        'Technology': front_office_directory[2022][6],
        'Ticket Operations': front_office_directory[2022][7],
        'Communications & Broadcasting': front_office_directory[2022][8],
        'Marketing & Business Development': front_office_directory[2022][9],
        'Corporate Partnerships & Client Services': front_office_directory[2022][10],
        'Ticket Sales & Services': front_office_directory[2022][11],
        'Ballpark Operations & Development': front_office_directory[2022][12],
        'Event Operations & Guest Experience': front_office_directory[2022][13],
        'Groundskeeping & Landscaping': front_office_directory[2022][14],
        'Community Impact': front_office_directory[2022][15],
        'Major League Baseball': front_office_directory[2022][16]
    }
}

for year in range(min(_source.keys()), max(_source.keys()) + 1):
    for k in employees[year]:
        employees[year][k]['Year'] = year
        employees[year][k]['Department'] = k
        employees[year][k]['Source of Information'] = _source[year]
        final_df = pd.concat((final_df, employees[year][k]))

final_df.dropna(subset=['Name', 'Title'], inplace=True)
```

```python
final_df['Team'] = team
final_df['NameLast'] = final_df.Name.apply(lambda x: HumanName(x).last)
final_df['NameFirst'] = final_df.Name.apply(lambda x: HumanName(x).first)

final_df[['Year', 'Team', 'NameLast', 'NameFirst', 'Title', 'Department', 'Source of Information']]
```

```python
final_df[['Year', 'Team', 'NameLast', 'NameFirst', 'Title', 'Department', 'Source of Information']] \
    .to_excel(f'../out/{team}-{min(_source.keys())}-to-{max(_source.keys())}.xlsx', index=False)
```
